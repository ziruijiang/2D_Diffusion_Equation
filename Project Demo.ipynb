```python
import numpy as np
from DE_solver import DE_solver
import DE_Plotter
%matplotlib inline
```

## Case 1: Uniorm Mesh

```python
x_pos = np.array([0,4,8,12,16])
y_pos = np.array([0,4,8,12,16])
D_mesh = np.array([[1,0.25,1,0.25],
                  [0.25,1,0.25,1],
                  [1,0.25,1,0.25],
                  [0.25,1,0.25,1]])
abs_mesh = np.array([[0.5,0.2,0.5,0.2],
                  [0.2,0.5,0.2,0.5],
                  [0.5,0.2,0.5,0.2],
                  [0.2,0.5,0.2,0.5]])
source_mesh = np.array([[1,5,9,10],
                  [1,3,5,9],
                  [0.5,2,4,8],
                  [0.3,1,3,7]])
err_tol = 10**(-6)
```

```python
flux_gs,j,s = DE_solver(x_pos, y_pos, D_mesh, abs_mesh, source_mesh, err_tol)
```

```python
DE_Plotter.input_plot(x_pos, y_pos, source_mesh)
```

```python
%matplotlib notebook
%matplotlib notebook
DE_Plotter.output_plot(x_pos, y_pos, flux_gs)
```
